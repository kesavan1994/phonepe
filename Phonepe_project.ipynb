```python
# Create Data frame for Aggregated\ Transation

import os
import json
import pandas as pd

path="C:\\Users\\91909\\firstprogram\\Phonepe\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\"


agg_tran_st=os.listdir(path)
agg_tran_st

ats={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}
for i in agg_tran_st:
  tr_st=path+i+'/'
  agg_year=os.listdir(tr_st)
  for j in agg_year:
    tr_yr=tr_st+j+'/'
    agg_yr_list=os.listdir(tr_yr)
    for k in agg_yr_list:
      tr_j=tr_yr+k
      data=open(tr_j,'r')
      d=json.load(data)
      for z in d['data']['transactionData']:
        Name=z['name']
        Count=z['paymentInstruments'][0]['count']
        Amount=z['paymentInstruments'][0]['amount']
        ats['State'].append(i)
        ats['Year'].append(j)
        ats['Quater'].append(int(k.strip('.json')))
        ats['Transacion_amount'].append(Amount)
        ats['Transacion_count'].append(Count)
        ats['Transacion_type'].append(Name)

ats=pd.DataFrame(ats)      
ats['State']=ats['State'].str.replace('-',' ')
ats['Transacion_type']=ats['Transacion_type'].str.replace('-',' ')

ats
```

```python
import locale
from math import log10
def sexify(value):
    MAP = {
            0: '',
            1: '',
            2: '',
            3: 'Thousand',
            4: 'Thousand',
            5: 'Lakh',
            6: 'Lakh',
            7: 'Crore',
            8: 'Crore',
        }
    power = int(log10(abs(value)))
    if power <= 8:

        humanized = locale.format('%d', value, grouping=True)
        rounded = humanized[:4].replace(',', '.')
        s = rounded + ' ' + MAP[power]
        if (rounded != '1.00'): s = s + 's'
    else:
        value = value / 10000000
        humanized = locale.format('%d', value, grouping=True)
        s = humanized + ' ' + 'Crores'

    print(s)
sexify(3984674.4287863)   
```
